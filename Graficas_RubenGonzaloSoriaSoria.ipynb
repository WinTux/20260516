# Importando librerías

```python
import matplotlib.pyplot as plt
```

# Ejecutando el código Figura 12

```python
competencias = [
    'Pensamiento crítico',
    'Ética digital',
    'Comprensión tecnológica',
    'Uso técnico'
]

valores = [40, 25, 20, 15]

plt.figure(figsize=(8,5))
plt.bar(competencias, valores)

plt.xlabel('Competencias')
plt.ylabel('Nivel de importancia')
plt.title('Competencias para alfabetización en IA')

plt.show()
```

# Ejecutando Figura 14

```python
areas = [
    'Protección de datos',
    'Ética algorítmica',
    'Transparencia tecnológica',
    'Responsabilidad profesional',
    'Inclusión digital'
]

valores = [30, 25, 20, 15, 10]

plt.figure(figsize=(9,5))
plt.bar(areas, valores)

plt.xlabel('Áreas de formación ética')
plt.ylabel('Porcentaje de relevancia')
plt.title('Áreas prioritarias de formación ética en Ingeniería de Sistemas')

plt.xticks(rotation=10)

plt.show()
```
