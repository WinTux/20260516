# Importando librerías

```python
import matplotlib.pyplot as plt
```

# Ejecutando el código

```python
competencias = [
    'Pensamiento crítico',
    'Ética digital',
    'Comprensión tecnológica',
    'Uso técnico'
]

valores = [40, 25, 20, 15]

plt.figure(figsize=(8,5))
plt.bar(competencias, valores)

plt.xlabel('Competencias')
plt.ylabel('Nivel de importancia')
plt.title('Competencias para alfabetización en IA')

plt.show()
```
